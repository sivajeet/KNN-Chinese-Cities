### Find the best K

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=4, test_size=0.3)
scores = []
for i in range(10,30):
    c = Classifier(n_neighbors=i, metric="hamming")
    c.fit(X_train, y_train)
    prediction = c.predict(X_test)
    score =c.score(X_test, y_test)
    scores.append((i, score))
    print("n_neighbors: "+ str(i))
```

### Plot scores

```python
i=  [i for i,score in scores ]
score_values=  [score for i,score in scores ]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(i, score_values)
plt.xlabel('Value of K for KNN')
plt.ylabel('Score')
```

it seems 24 is the best one
